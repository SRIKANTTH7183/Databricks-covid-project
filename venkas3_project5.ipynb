```python
import pandas as pd

file_path = "/Volumes/workspace/default/covid_data/OWID_COVID19_data_4_Project5.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.rename(columns={
    "people_fully_vaccinated_phun": "people_fully_vaccinated_per_hundred",
    "new_cases_pmil": "new_cases_per_million"
}, inplace=True)

df.columns
```

```python
df = df[
    (df["continent"].notna()) &
    (df["people_fully_vaccinated_per_hundred"].notna()) &
    (df["new_cases_per_million"].notna()) &
    (df["people_fully_vaccinated_per_hundred"] > 0) &
    (df["new_cases_per_million"] > 0)
]

continent_counts = df.groupby("continent").size().reset_index(name="count")
continent_counts
```

```python
df["date"] = pd.to_datetime(df["date"])

df["month"] = df["date"].dt.month
df["year"] = df["date"].dt.year

df = df[(df["year"] == 2021) & (df["month"].isin([9, 10]))]

print("Total record count:", len(df))
df.head()
```

```python
continent_month_counts = df.groupby(["continent", "month"]).size().reset_index(name="count")
continent_month_counts
```

```python
avg_df = df.groupby(["continent", "month"]).agg({
    "people_fully_vaccinated_per_hundred": "mean",
    "new_cases_per_million": "mean",
    "excess_mortality": "mean"
}).reset_index()

avg_df
```

```python
avg_df = avg_df.dropna()
```

```python
import matplotlib.pyplot as plt

avg_df.plot(
    kind="bar",
    x="continent",
    y=["people_fully_vaccinated_per_hundred", "new_cases_per_million"],
    color=["red", "blue"]   # 👈 change colors here
)

plt.title("Vaccination vs COVID Cases by Continent (Sep-Oct 2021)")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.show()
```

```python
corr_cases = avg_df["people_fully_vaccinated_per_hundred"].corr(avg_df["new_cases_per_million"])
corr_mortality = avg_df["people_fully_vaccinated_per_hundred"].corr(avg_df["excess_mortality"])

print("Vaccination vs Cases:", corr_cases)
print("Vaccination vs Mortality:", corr_mortality)
```

```python
analysis_df = df[
    (df["people_fully_vaccinated_per_hundred"].notna()) &
    (df["new_cases_per_million"].notna()) &
    (df["people_fully_vaccinated_per_hundred"] > 0) &
    (df["new_cases_per_million"] > 0) &
    (df["year"] == 2021) &
    (df["month"].isin([9, 10])) &
    (df["continent"].notna())
].copy()
```

```python
missing_gdp = analysis_df[analysis_df["gdp_per_capita"].isna()]["location"].drop_duplicates()
missing_gdp
```

```python
gdp_fill_data = [
    ("Cuba", 9500),
    ("Liechtenstein", 139000),
    ("Monaco", 115000),
    ("Somalia", 1500),
    ("Syria", 3000),
    ("Taiwan", 73000),
    ("Afghanistan", 2200),
    ("Yemen", 2500),
    ("Sudan", 4000),
    ("South Sudan", 1500),
    ("Eritrea", 2000),
    ("Zimbabwe", 2500),
    ("Haiti", 3000),
    ("Nepal", 4000)
]
gdp_fill_df = pd.DataFrame(gdp_fill_data, columns=["location", "gdp_fill"])
gdp_fill_df
```

```python
analysis_df = analysis_df.merge(gdp_fill_df, on="location", how="left")

analysis_df["gdp_per_capita"] = analysis_df["gdp_per_capita"].fillna(analysis_df["gdp_fill"])

analysis_df.drop(columns=["gdp_fill"], inplace=True)
```
